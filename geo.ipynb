```python
#1.6: example on a missy code and how to clean it 

geothermal_grad1 = 0.033
temp_f1 =123
depth_meters1=((temp_f1 -32)*5/9)/geothermal_grad1
geothermal_grad2 = 0.012
temp_f2 = 230
depth_meters2=((temp_f2 -32)*5/9)/geothermal_grad2
geothermal_grad3 = 0.065
temp_f3 =450
depth_meters3=((temp_f3 -32)*5/9)/geothermal_grad3
print(depth_meters1, depth_meters2 , depth_meters3 )
```

```python
# section 1.6 how to clean the missy code in the example above, the steps, this 
# was a multiple step process that i reorganized after following his steps. 

# step1: assign a function of the mathmatical relationships in the original code. 
def fahrenheit_to_celsius(temp_f):
    return((temp_f-32)*5/9)
#print(fahrenheit_to_celsius)
def temp_to_depth (temp_celsius,geotherm_gradient):
    return(temp_celsius/geotherm_gradient)
#print(temp_to_depth)
    
# step 2: arrange all similar variables under each other
geothermal_grad1 = 0.033
geothermal_grad2 = 0.012
geothermal_grad3 = 0.065

temp_f1 =123
temp_f2 = 230
temp_f3 =450

# step 3: define a new variable (data frame) called data 1 ( called tuple)
data1= (fahrenheit_to_celsius(temp_f1),geothermal_grad1)
data2= (fahrenheit_to_celsius(temp_f2),geothermal_grad2)
data3= (fahrenheit_to_celsius(temp_f3),geothermal_grad3)

# step 4: create a list of all the data. Remember: use [  ] and called it all_data
all_data = [data1, data2, data3]

# print 
for data in all_data: 
    print(temp_to_depth (data[0],data[1]))
```
